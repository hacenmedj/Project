```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris= pd.read_csv("C:/Users/hotsa/Downloads/iris/iris.csv")
iris
```

```python
iris.variety.value_counts()
```

```python
#transformation des noms des noms de variétés en nombres
iris["variety"] = iris["variety"].replace({"Setosa": 0, "Versicolor": 1,"Virginica": 2})
iris
```

```python
#suppression de la colonne variété
iris1 = iris.drop("variety", axis=1)
iris1
```

```python
#caracteristiques de chaques plantes et creation d'une liste
datairis = iris1.values.tolist()
print(datairis)
```

```python
#type de fleurs
iris2= iris.variety.values.tolist()
iris2
```

```python
clf=RandomForestClassifier()
clf.fit(datairis,iris2)
```

```python
# exemple de prediction pour une fleur
prediction=clf.predict([[5,2,4,3]])
print(prediction)
```
